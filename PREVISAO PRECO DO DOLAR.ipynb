```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pickle
tf.__version__
```

```python
df_dolar_brl = pd.read_csv("H:/Barbosa/Downloads/Month.csv", index_col=False)
```

```python
df_dolar_brl
```

```python
date_array = np.array(df_dolar_brl["Date"])
date_array
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_sequences(date_array)
tokenizer.fit_on_texts(date_array)
```

```python
date_array_text = tokenizer.texts_to_sequences(date_array)
```

```python
date_array_text
```

```python
df_dolar_brl
```

```python
selected_columns = ['Last', 'Opening']
df_dolar_brl = df_dolar_brl[selected_columns]
```

```python
X_train = np.array(date_array_text)
y_train = df_dolar_brl['Opening']
```

```python
X_train.shape
```

```python
y_train = np.array(y_train)
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
x_scaler = MinMaxScaler()
y_scaler = MinMaxScaler()
```

```python
X_train = x_scaler.fit_transform(X_train)
y_train = y_scaler.fit_transform(y_train.reshape(-1,1))
```

```python
X_train
```

```python
y_train
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=332, output_dim=32, input_length=3),
    tf.keras.layers.LSTM(units=10),
    tf.keras.layers.Dense(units=1, activation='linear')
])
```

```python
model.compile(optimizer='Adam', loss='mean_squared_error')
```

```python
model.fit(X_train, y_train, epochs=500)
```

```python
test_date = ['12/08/2023']
tokenizer.fit_on_sequences(test_date)
tokenizer.fit_on_texts(test_date)
test_date = tokenizer.texts_to_sequences(test_date)
test_date_normalized = x_scaler.transform(test_date)
test_date_normalized
```

```python
predict = model.predict(test_date_normalized)
```

```python
predict
```

```python
predict = predict.reshape(1,-1)
```

```python
predict
```

```python
predict_new = y_scaler.inverse_transform(predict)
```

```python
predict_new
```

```python
with open('tokenizer.p', 'wb') as filetokenizer:
    pickle.dump(tokenizer, filetokenizer)
    
with open('x_scaler.p', 'wb') as x_fileminmax:
    pickle.dump(x_scaler, x_fileminmax)
    
with open('y_scaler.p', 'wb') as y_fileminmax:
    pickle.dump(y_scaler, y_fileminmax)

model.save("model.h5")
```
